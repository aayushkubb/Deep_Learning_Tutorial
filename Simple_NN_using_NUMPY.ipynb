```python
# Features
x1=2
x2=5
x3=3

# Target Value
y=100

# Learning Rate
lr=0.01

#epoch
epoch=100
```

```python
# Randomly initialize Weights
w1=0
w2=0
w3=0
```

For loop which will run 100 times

prediction=multiply each input with its weight

Calculate error -> Squared error

Adjust/Update the weights by calculating derivative of error with respect to weight

w_new = w_old - lr(partial derivative of ERROR wrt w_old)




```python
for i in range(epoch):
  # Forward Pass
  y_pred=x1*w1 + x2*w2 + x3*w3

  #Calculate Squared Error
  error=(y-y_pred)**2

  # Update Weights (BP)
  # dERROR/dw -> 2*((y-y_pred))*d(x1*w1 + x2*w2 + x3*w3)/dw
  # dERROR/dw1 -> 2*((y-y_pred))*(-x1)

  dEw1=2*(y-y_pred)*(-x1)
  dEw2=2*(y-y_pred)*(-x2)
  dEw3=2*(y-y_pred)*(-x3)

  print("Epoch: ",i)
  print("Weights: w1:",w1," ,w2: ",w2," ,w3: ",w3)
  print("Error: ",error)
  print("_"*10)

  w1=w1- lr*dEw1
  w2=w2- lr*dEw2
  w3=w3- lr*dEw3

```

```python


import numpy as np

# Features
x1 = 2
x2 = 5
x3 = 3

# Target Value
y = 100

# Learning Rate
lr = 0.01

# Epochs
epoch = 100

# Initialize Weights
w1 = 0.0
w2 = 0.0
w3 = 0.0

# Sigmoid Activation
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Derivative of Sigmoid
def sigmoid_derivative(a):
    return a * (1 - a)

for i in range(epoch):
    # ---------------------------
    # Forward Pass
    # ---------------------------
    z = w1*x1 + w2*x2 + w3*x3      # Linear combination
    a = sigmoid(z)                 # Activation output

    # Compute prediction
    y_pred = a                     # Use activated value

    # Compute Error (MSE for 1 sample)
    error = (y - y_pred)**2

    # ---------------------------
    # Backpropagation
    # ---------------------------
    # dL/da
    dL_da = 2*(y_pred - y)         # derivative of (y_pred - y)^2

    # da/dz for sigmoid
    da_dz = sigmoid_derivative(a)

    # chain rule: dL/dz
    dL_dz = dL_da * da_dz

    # Gradients for each weight
    dL_dw1 = dL_dz * x1
    dL_dw2 = dL_dz * x2
    dL_dw3 = dL_dz * x3

    # ---------------------------
    # Update Weights
    # ---------------------------
    w1 = w1 - lr*dL_dw1
    w2 = w2 - lr*dL_dw2
    w3 = w3 - lr*dL_dw3

    # Print progress occasionally
    if i % 10 == 0:
        print(f"Epoch {i}, Error: {error:.4f}, Prediction: {y_pred:.4f}")

print("\nFinal Weights:")
print("w1:", w1)
print("w2:", w2)
print("w3:", w3)
print("Final Prediction:", sigmoid(w1*x1 + w2*x2 + w3*x3))
```
