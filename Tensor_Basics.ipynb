```python
import tensorflow as tf

# 0D Tensor (scalar)
t0 = tf.constant(7)

# 1D Tensor (vector)
t1 = tf.constant([1, 2, 3])

# 2D Tensor (matrix)
t2 = tf.constant([[1, 2],
                  [3, 4]])

# 3D Tensor (cube / stack of matrices)
t3 = tf.constant([[[1],[2]], [[3],[4]]])

print(t0.shape, t1.shape, t2.shape, t3.shape)
```

```python
x = tf.constant([[1.0, 2.0],
                 [3.0, 4.0]])

y = tf.constant([[10.0, 20.0],
                 [30.0, 40.0]])
```

```python


# Element-wise operations
print("Add:\n", x + y)
print("Multiply:\n", x * y)
print("Subtract:\n", x - y)
```

```python
print("Add:\n", x + 10)
```

```python

# Matrix multiplication
print("MatMul:\n", tf.matmul(x, y))
```

```python

# Reductions
print("Sum:", tf.reduce_sum(x))
print("Mean:", tf.reduce_mean(x))

# Reshape
print("Reshape:\n", tf.reshape(x, (4,)))
```

```python
import tensorflow as tf

# Training data (X=inputs, y=targets)
X = tf.constant([[2.0], [4.0], [6.0], [8.0]])
y = tf.constant([[4.0], [8.0], [12.0], [16.0]])

# Trainable parameters (weights & bias)
w = tf.Variable(0.0)
b = tf.Variable(0.0)

lr = 0.01
epochs = 200

for epoch in range(epochs):
    with tf.GradientTape() as tape:
        y_pred = X * w + b
        loss = tf.reduce_mean((y - y_pred)**2)

    dw, db = tape.gradient(loss, [w, b])

    w.assign_sub(lr * dw)
    b.assign_sub(lr * db)

    if epoch % 40 == 0:
        print(f"Epoch {epoch}, Loss={loss.numpy():.4f}, w={w.numpy():.4f}, b={b.numpy():.4f}")

print("\nFinal parameters:")
print("w =", w.numpy(), " b =", b.numpy())
print("Prediction for X=10:", (10*w + b).numpy())
```
